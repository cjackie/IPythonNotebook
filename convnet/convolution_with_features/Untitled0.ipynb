```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
# feature_path, and target_image_path
feature_paths = ['features/edge1.png', 'features/edge2.png', 'features/edge3.png', 
                 'features/edge4.png']
target_image_path = 'images/me.png'
```

```python
# all images are in 8-bit pixels, black and white. m x n shape

# open features
feature_shape = (20,20)
features = {}
for path in feature_paths:
    f = Image.open(path);
    f = f.resize(feature_shape, resample=Image.NEAREST)
    f = f.convert('L')  # make the image 8-bit pixels, black and white
    f = mpimg.pil_to_array(f)  # make it into a ndarray
    f = f*(1/255.0)  # scale it
#     # tranform
#     for i in range(feature_shape[0]):
#         for j in range(feature_shape[1]):
#             f[i,j] = 1.0 if f[i,j] > 0.5 else 0.0
    features[path] = f
    
# open target images
target_image = Image.open(target_image_path)
target_image = target_image.convert('L')
target_image = mpimg.pil_to_array(target_image)


```

```python
# plot featuers and target_image
features_fig = plt.figure()
for path, i in zip(feature_paths, range(len(feature_paths))):
    features_fig.add_subplot(1, len(feature_paths), i+1)
    plt.imshow(features[path], cmap='gray')
    
target_image_figure = plt.figure()
plt.imshow(target_image, cmap='gray')
```

```python
import tensorflow as tf

img_ph = tf.placeholder(tf.float32, shape=(1,)+target_image.shape+(1,), name='img_ph')
filter_ph = tf.placeholder(tf.float32, feature_shape+(1,1), name='filter_ph')
# tf.nn.convolution(input, filter, padding, strides=None, dilation_rate=None, name=None, data_format=None)
# see https://www.tensorflow.org/api_docs/python/nn/convolution
conv_op = tf.nn.convolution(img_ph, filter_ph, 'SAME', strides =(1,1))
conv_op
```

```python
# open a session 
sess = tf.Session()

# prepare data
img_in = np.ones(shape=img_ph.get_shape())
img_in[0,:,:,0] = target_image

filter_in = np.ones(shape=filter_ph.get_shape())
filter_in[:,:,0,0] = features[feature_paths[2]]

o = sess.run(conv_op, feed_dict={img_ph: img_in, filter_ph: filter_in})
plt.imshow(o[0,:,:,0], cmap='gray')

```
