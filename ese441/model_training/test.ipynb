```python
from motion_basis_learning import MotionBasisLearner
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from descriptor_extractor import DescriptorExtractor

import cProfile

%matplotlib inline
```

```python
accel_data1 = []
accel_file1 = open('../data/accelerometer_1.csv', 'r')
accel_file1.readline()
for line in accel_file1:
    accel_data1.append(map(lambda s: float(s), line.split(',')))
accel_file1.close()
accel_data2 = []
accel_file2 = open('../data/accelerometer_2.csv', 'r')
accel_file2.readline()
for line in accel_file2:
    accel_data2.append(map(lambda s: float(s), line.split(',')))
accel_file2.close()

shrink_size = 500
accel_data = np.stack([np.transpose(np.array(accel_data1))[0:3,0:shrink_size],
                      np.transpose(np.array(accel_data2))[0:3,0:shrink_size]])
accel_data = np.expand_dims(accel_data, -1)
time_axis = range(shrink_size)
```

```python
basis_learner = MotionBasisLearner()
# c.build_training_model(accel_data, restore_params_path=MotionBasisLearner.SAVE_PARAMS_PATH+'-10')
basis_learner.build_training_model(accel_data)
```

```python
pr = cProfile.Profile()
# pr.enable()
basis_learner.train(verbose=True, steps=11, learning_rate=0.01 )
# pr.disable()
# pr.print_stats()
```

```python
d = basis_learner._gen_v(tf.convert_to_tensor(basis_learner._gen_h(tf.convert_to_tensor(accel_data, tf.float32)), tf.float32), 0.001)
# plt.plot(range(d.shape[2]), d[0,1,:,0])
# plt.plot(range(accel_data.shape[2]), accel_data[0,1,:,0])

plt.plot( d[0,1,200:300,0])
plt.plot(accel_data[0,1,200:300,0])
```

```python
descriptor_extractor = DescriptorExtractor(basis_learner)
```

```python
d = descriptor_extractor.extract_descriptor(accel_data[0,::,:,0])
d_raw = descriptor_extractor._extract_descriptor_raw(accel_data[0,:,:,0])
```
