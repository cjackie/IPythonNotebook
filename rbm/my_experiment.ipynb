```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf
import time
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

%matplotlib inline
```

```python
images = mnist.train.images
batch_size, img_vec_size = images.shape[0], images.shape[1]
img_dim = int(math.sqrt(img_vec_size))
```

```python
# model
hidden_units_num = 500
# right now, only one w...
with tf.variable_scope('rbm') as rbm_scope:
#     rbm_scope.reuse_variables()
    # ignore biases terms for visible units and hidden units for now.
    w = tf.get_variable('weights', shape=(img_vec_size, hidden_units_num), dtype=tf.float32,
                        initializer=tf.random_normal_initializer(stddev=1))
    hb = tf.get_variable('hidden_biases', shape=(hidden_units_num,), dtype=tf.float32,
                         initializer=tf.random_normal_initializer(stddev=1))
    vb = tf.get_variable('visible_biases', shape=(img_vec_size,), dtype=tf.float32,
                         initializer=tf.random_normal_initializer(stddev=1))

# probabilities
v_in = tf.placeholder(tf.float32, shape=(None, img_vec_size))
h_in = tf.placeholder(tf.float32, shape=(None, hidden_units_num))
p_h = tf.nn.sigmoid(tf.matmul(v_in, w) + hb) # maybe?
p_v = tf.nn.sigmoid(tf.matmul(h_in, tf.transpose(w)) + vb)

# error
v_in2 = tf.placeholder(tf.float32, shape=(None, img_vec_size))
error = tf.reduce_mean(tf.square(v_in - v_in2))

sess = tf.Session()
sess.as_default()
sess.run(tf.global_variables_initializer())
```

```python
# learning rate
learning_rate = 0.01
images_num = 1
start_image_index = 2

# fit data
v = images[start_image_index:start_image_index+images_num, :]
v_real = np.array(v).reshape((images_num,img_vec_size))
for i in range(images_num):
    for j in range(img_vec_size):
        # compress to each 1 or 0 for each pixels
        v_real[i,j] = 1 if v_real[i,j] > 0 else 0
        
training_steps = 5
for _ in range(training_steps):
    # gibbs sampling
    # sample hidden layer
    hprobs = sess.run(p_h, feed_dict={v_in: v_real})
    h_real = np.ndarray((hprobs.shape))
    for i in range(h_real.shape[0]):
        for j in range(h_real.shape[1]):
            h_real[i,j] = 1 if np.random.rand() < hprobs[i,j] else 0
            
    # sample reconstruction
    steps = 2
    h_fantasy = h_real.copy()
    for step in range(steps):
        vprobs = sess.run(p_v, feed_dict={h_in: h_fantasy})
        v_fantasy = np.ndarray((vprobs.shape))
        for i in range(vprobs.shape[0]):
            for j in range(vprobs.shape[1]):
                v_fantasy[i,j] = 1 if np.random.rand() < vprobs[i,j] else 0
        hprobs = sess.run(p_h, feed_dict={v_in: v_fantasy})
        for i in range(hprobs.shape[0]):
            for j in range(hprobs.shape[1]):
                h_fantasy[i,j] = 1 if np.random.rand() < hprobs[i,j] else 0
            
    # # update weights
    delta_w = ((1.0/images_num)*np.dot(np.transpose(v_real), h_real)
        - (1.0/images_num)*np.dot(np.transpose(v_fantasy), h_fantasy))*learning_rate
    delta_vb = (1.0/images_num)*np.sum((v_real - v_fantasy), axis=0)*learning_rate
    delta_hb = (1.0/images_num)*np.sum((h_real - h_fantasy), axis=0)*learning_rate
    sess.run(w.assign_add(delta_w))
    sess.run(vb.assign_add(delta_vb))
    sess.run(hb.assign_add(delta_hb))

# for i in range(w.get_shape()[0]):
#     for j in range(w.get_shape()[1]):
#         np.sum(h_fantasy, axis=0)
#         w[i,j].assign_add(


feed_dict = {
    v_in: v_real,
    v_in2: v_fantasy
}
print(sess.run(error, feed_dict=feed_dict))
plt.imshow(sess.run(w)[:,1].reshape((img_dim, img_dim)), cmap='gray')
```
